    <h2>Difference between <code>iloc</code> and <code>loc</code> in Pandas</h2>
    <table>
        <tr>
            <th>Feature</th>
            <th>iloc</th>
            <th>loc</th>
        </tr>
        <tr>
            <td>Indexing Type</td>
            <td>Integer-based</td>
            <td>Label-based</td>
        </tr>
        <tr>
            <td>Syntax</td>
            <td><code>df.iloc[row_index, col_index]</code></td>
            <td><code>df.loc[row_label, col_label]</code></td>
        </tr>
        <tr>
            <td>Supports Slicing</td>
            <td>Yes (integer indices)</td>
            <td>Yes (label-based)</td>
        </tr>
        <tr>
            <td>Works with Boolean Indexing</td>
            <td>No</td>
            <td>Yes</td>
        </tr>
    </table>
    
<h3> Example: </h3>
<pre>
import pandas as pd

data = {'A': [10, 20, 30], 'B': [40, 50, 60]}
df = pd.DataFrame(data, index=['x', 'y', 'z'])

### Using iloc (integer positions)
print(df.iloc[1, 0])  # Output: 20 (row index 1, column index 0)

### Using loc (label-based)
print(df.loc['y', 'A'])  # Output: 20 (row 'y', column 'A')
